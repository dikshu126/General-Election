```python
from sklearn.preprocessing import MinMaxScaler

df=pd.read_csv('/content/Indian Candidates for General Election.csv')
data = df[['WINNER','TOTAL ELECTORS','AGE']]
scaler = MinMaxScaler()

# Fit and transform the data
scaled_data = scaler.fit_transform(data)

print("Original Data:")
print(data)

# Display the scaled data
print('\n',"Scaled Data:")
print(scaled_data)
```

```python
df=pd.read_csv('/content/Indian Candidates for General Election.csv')

# 1. Filtering Data
filtered_data = df[df['AGE'] <20]
print("1. Filtered Data (AGE < 20):")
print(filtered_data)
print()

# 2. Sorting Data
sorted_data = df.sort_values(by='AGE', ascending=False)
print('\n',"2. Sorted Data (by AGE in descending order):")
print(sorted_data)
print()

# 3. Grouping and Aggregating Data
grouped_data = df.groupby('GENDER').agg({'AGE': 'mean'}).reset_index()
print('\n',"3. Grouped and Aggregated Data (Average AGE by Gender):")
print(grouped_data)
print()

# 4. Adding or Modifying Columns
df['Senior'] = df['AGE'] > 60
print('\n',"4. Data with Added Column 'Senior':")
print(df)
print()

df.to_csv('transformed_data.csv', index=False)
```

```python
#let plot a bar chart in EDUCATIONAL QUALIFICATION of MP's
df=pd.read_csv('/content/Indian Candidates for General Election.csv')
plt.figure(figsize=(14, 6))
df['EDUCATION'].value_counts().plot(kind='bar',color=['Blue','Orange','Yellow','Green','Maroon','pink','skyblue','Gray','Green','red','salmon','black'])
plt.title('Distribution of Educational Qualifications of Elected MPs')
plt.xlabel('Education Level')
plt.ylabel('Number of MPs')
plt.xticks(rotation=45)
plt.show()
```
